```
tips=sns.load_dataset("tips")
tips.head()
```

```
tips.shape
```

Count plot of day

```
sns.countplot(tips.day)
plt.show()
```

```
sns.countplot(tips.day,palette=["b","r","g","yellow"])
plt.show()
```

```
sns.countplot(tips.sex)
plt.show()
```

```
#how we cam check how many smokers are in this dataset?
sns.countplot(tips.smoker)
plt.show()
#sns.countplot(x=smoker,data=tips))
```

Barplot with tip by day

```
sns.barplot(x='day',y='tip',data=tips)
plt.show()
```

```
sns.barplot(x='day',y='tip',ci=None,data=tips)
plt.show()
```

```
sns.set(rc={'figure.figsize':(12,8)})#change size
sns.barplot(x='day',y='tip',ci=None,data=tips)
plt.show()
```

```
sns.reset_orig()
sns.barplot(x='day',y='tip',ci=None,data=tips)
plt.show()
```

```
plt.figure(figsize=(20,10))#change size for one time
sns.barplot(x='day',y='tip',ci=None,data=tips)
plt.show()
```

```
sns.set_style('darkgrid')
sns.set_palette("RdBu")
sns.set_context("poster")

sns.barplot(x='day',y='tip',ci=None,data=tips)
plt.show()
```

```
sns.reset_orig()
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=tips)
plt.show()
```

```
sns.scatterplot(x='total_bill',y='tip',hue='smoker',hue_order=['Yes','No'],data=tips)
plt.show()
```

```
plt.figure(figsize=(20,10))
sns.scatterplot(x='total_bill',y='tip',hue='smoker',hue_order=['Yes','No'],
                size='day',style='sex',data=tips)
plt.show()
```

```
sns.regplot(x='total_bill',y='tip',data=tips,
            kind='scatter',
            col='smoker',
            row='time')
plt.show()
```

```
import pandas as pd
path='https://raw.githubusercontent.com/RamiKrispin/coronavirus/master/csv/coronavirus.csv'

df=pd.read_csv(path)
df.head()
```

```
df.shape
```

```
df.info()
```

```
sns.countplot(x=df.type)
plt.show()
```

```
sns.barplot(x='country',y='cases',ci=None,data=df)
plt.xticks(rotation=90)
plt.show()
```

```
sns.set(rc={'figure.figsize':(40,20)})
sns.barplot(x='country',y='cases',ci=None,data=df)
plt.xticks(rotation=90)
plt.show()
```

```
df1=df[(df['country']=='US')|(df['country']=='Russia')|(df['country']=='Brazil')]
df1.shape
```

```
df1.head()
```

```
sns.lineplot(x='date',y='cases',hue='country',ci=None,data=df1)
plt.xticks(rotation=90)
plt.show()
```

```
sns.relplot(x='date',y='cases',row='country',
            height=5,
            aspect=5,)
```
